# **5 Построить модель, которая предсказывает оценку для фильма и пользователя (воспользоваться RandomForestRegressor, тк мы решаем задачу регрессии)**

```
data = data.drop('timestamp', axis=1)
data.head()
```

```
x = data.drop('rating', axis = 1)
y = data['rating']
```

```
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 50, random_state = 0)
```

```
regressor.fit(X_train, y_train)
```

```
y_pred = regressor.predict(X_test)
```

# **6 Оценить качество регрессора на отложенной (train_test_split) выборке по одной из метрик на выбор - RMSE, RMAE, R^2**

```
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# ** 10 Средствами спарка соедините информацию по фильмам и жанрам (u.genre) Дублирование строк в данном случае - ок.Интересует такая структура - id фильма | название фильма | жанр**



```
all_column_names = item.columns
```

```
columns_to_remove = ['movie_id',
 'movie_title',
 'release_date',
 'video_release_date',
 'IMDb_URL']
```

```
columns_for_replacement = [i for i in all_column_names if i not in columns_to_remove]
columns_for_replacement
```

```
from pyspark.sql.functions import col, when
for i in columns_for_replacement:
    item = item.withColumn(i,when((col(i)==1), i).otherwise(col(i)))
item.show()
```

```
from pyspark.sql.functions import col, when
for i in columns_for_replacement:
    item = item.withColumn(i,when((col(i)==0), None).otherwise(col(i)))
item.show()
```

```
item1 = item.drop('release_date', 'video_release_date', 'IMDb_URL', 'movie_title')
item1.show()
```

```
import pyspark.sql.functions as f
item1 = item1.withColumn('genre', f.concat_ws(',', *[c for c in item1.columns if c!='movie_id']))
```

```
item1.show()
```

```
item3 = item.select('movie_id', 'movie_title')
item3.show()
```

```
item4 = item1.select('movie_id', 'genre')
item4.show()
```

```
films_and_genre_join = item3.join(item4, on='movie_id', how='left')
films_and_genre_join.show()
```

```
from pyspark.sql.functions import col, explode, regexp_replace, split

films = films_and_genre_join.withColumn(
    "genre", 
    explode(split(col("genre"), ","))
)
films.show()
```
