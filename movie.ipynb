```python
import pandas as pd
```

### 1 Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

```python
ratings = pd.read_csv('u.data.csv', sep="\t", names = ["user_id", "item_id", "rating", "timestamp"] )
```

```python
movies = pd.read_csv('u.item.csv', encoding = "ISO-8859-1", sep="|", names = ["movie_id", "movie_title", "release_date", "video_release_date", "IMDb_URL", "unknown", "Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"] )
```

```python
ratings.head()
```

```python
movies.head()
```

### 2 Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```python
ratings.groupby('user_id')['rating'].sum().sort_values(ascending=False).head(1)
```

### 3 Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 

```python
ratings1 = ratings[ratings['user_id'] == 450]
```

```python
ratings1.head()
```

```python
ratings1.drop(['user_id', 'timestamp'], axis='columns', inplace=True)
```

```python
ratings1.head()
```

### 4 Добавьте к датафрейму из задания 3 столбцы:
По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет
Cтолбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей


```python
ratings_movies = pd.merge(ratings1, movies, how='left', left_on='item_id', right_on='movie_id')
```

```python
ratings_movies.head(5)
```

```python
ratings_movies.drop(['movie_id', 'movie_title', 'video_release_date','IMDb_URL','IMDb_URL'], axis='columns', inplace=True)
```

```python
ratings_movies.head()
```

```python
ratings_sum = ratings.groupby('item_id')['rating'].agg(['sum', 'count'])
```

```python
ratings_sum.head()
```

```python
movies1 = pd.merge(ratings_movies, ratings_sum, how='left', on='item_id')
```

```python
movies1.head()
```

```python
movies1['date'] = pd.to_datetime(movies1['release_date'], errors='coerce')
```

```python
movies1['release_year'] = movies1['date'].dt.year
```

```python
movies1.drop(['release_date', 'date'], axis='columns', inplace=True)
```

```python
movies1.head()
```

### 5 Сформируйте X_train, X_test, y_train, y_test

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
movies1.columns.values.tolist()
```

```python
X, Y = movies1[['unknown', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', 'sum', 'count', 'release_year']], movies1['rating']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

### 6 Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
X_train.head()
```

```python
Y_test.head(10)
```

```python
linear = LinearRegression()
linear.fit(X_train, Y_train)
linear.predict(X_test)
```

### 7 Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
mean_squared_error ( Y_test ,  linear.predict(X_test) ) 
```

```python
r2_score ( Y_test ,  linear.predict(X_test) ) 
```

```python
mean_absolute_error ( Y_test ,  linear.predict(X_test) ) 
```
