```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('1.csv', sep=';', comment='#', decimal=",")
data.head(10)
```

```python
data['Sum']=data['Quantity']*data['UnitPrice']
data.head(10)
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
data['Month'] = data['InvoiceDate'].dt.strftime('%m')
data.head(10)
```

```python
sales_dinamics = data.groupby(data['Month'])['Sum'].sum()
sales_dinamics
sales_dinamics.plot()
plt.grid()
plt.xlabel('Month')
plt.ylabel('Sales')
```

```python
country_sales = data.groupby('Country').nunique()[['InvoiceNo']].sort_values('InvoiceNo', ascending=False).reset_index()
country_sales.head(40)
```

```python
codes_sales = data.groupby(['StockCode']).sum()[['Sum']].sort_values('Sum', ascending=False)
codes_sales.head()
```

```python
codes_sales_g = data.loc[data['Country'] == 'Germany']
codes_sales_g_1 = codes_sales_g.groupby(['StockCode']).sum()[['Sum']].sort_values('Sum', ascending=False)
codes_sales_g_1.head(5)
```

```python
customer_sales = data.groupby('CustomerID').nunique()[['InvoiceNo']].sort_values('InvoiceNo', ascending=False)
print(customer_sales)
```
